# Figure 3 codes

This notebook contains the plotting code for main-text Figure 3a, Figure 3c, Figure 3d, Figure 3e, Figure 3f, and Figure 3g using the CSV files in `data/`.


```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

BASE_DIR = Path.cwd()
DATA_DIR = BASE_DIR / 'data'
OUTPUT_DIR = BASE_DIR / 'output'
OUTPUT_DIR.mkdir(exist_ok=True)

plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 25

TOPIC_COLORS = ['#5EA5C5', '#F08743', '#559F59']
TOPIC_ORDER = ['Politics', 'Gun_Control', 'Abortion']
TOPIC_TITLES = ['Partisan Alignment', 'Gun Control', 'Abortion Ban']
FOUR_COLORS = ['#E0E0E0'] + [sns.color_palette('ch:s=.25,rot=-.3', 10)[1]] + [sns.color_palette('ch:s=.25,rot=-.3', 10)[3]] + [sns.color_palette('ch:s=.25,rot=-.3', 10)[5]]
```

## Figure 3a


```python
homophily = pd.read_csv(DATA_DIR / 'figure3a_homophily.csv')
fig, ax = plt.subplots(1, 1, figsize=(25, 7.5))
markers = {'Politics': 'o', 'Gun_Control': '^', 'Abortion': 's'}
for topic, title, color in zip(TOPIC_ORDER, TOPIC_TITLES, TOPIC_COLORS):
    topic_df = homophily[homophily['topic'] == topic]
    sns.lineplot(data=topic_df, x='timestep', y='homophilic_mean', marker=markers[topic], markersize=8, ax=ax, label=title, color=color, errorbar=None)
ax.set_ylabel('Proprotion of Homophilic Interactions')
ax.set_xlabel('Time')
ax.legend(loc='lower right')
fig.savefig(OUTPUT_DIR / 'figure3a_homophily.pdf', bbox_inches='tight', pad_inches=0.1, dpi=300)
plt.show()
```

## Figure 3c


```python
effects = pd.read_csv(DATA_DIR / 'figure3cd_interaction_effects.csv')
same = effects[effects['interaction_type'] == 'same'].copy()
fig, ax = plt.subplots(1, 1, figsize=(10, 6.5))
sns.barplot(data=same, x='degree_bin', y='change_opinion', capsize=0.2, ax=ax, hue='topic_title', hue_order=TOPIC_TITLES, palette=TOPIC_COLORS, saturation=1)
ax.hlines(y=0, xmin=-0.5, xmax=2.5, lw=1, color='k')
ax.set_ylabel('Change in Polarization Level')
ax.set_xlabel('Types of Homophilic Interactions')
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['More Moderate', 'Same', 'More Radical'])
ax.set_xlim(-0.5, 2.5)
ax.legend()
fig.savefig(OUTPUT_DIR / 'figure3c_echo_chamber_effect_homophilic.pdf', bbox_inches='tight', pad_inches=0.1, dpi=300)
plt.show()
```

## Figure 3d


```python
diff = effects[effects['interaction_type'] == 'diff'].copy()
fig, ax = plt.subplots(1, 1, figsize=(10, 6.5))
sns.barplot(data=diff, x='degree_bin', y='change_opinion', capsize=0.2, ax=ax, hue='topic_title', hue_order=TOPIC_TITLES, palette=TOPIC_COLORS, saturation=1)
ax.hlines(y=0, xmin=-0.5, xmax=2.5, lw=1, color='k')
ax.set_ylabel('Change in Polarization Level')
ax.set_ylim(-0.6, 0.6)
ax.set_xlabel('Types of Opposing Interactions')
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['More Moderate', 'Same', 'More Radical'])
ax.set_xlim(-0.5, 2.5)
ax.legend(loc='lower left')
fig.savefig(OUTPUT_DIR / 'figure3d_echo_chamber_effect_opposing.pdf', bbox_inches='tight', pad_inches=0.1, dpi=300)
plt.show()
```

## Figure 3e-g


```python
bar_data = pd.read_csv(DATA_DIR / 'figure3efg_polarization_bar_data.csv')

def draw_polarization_bar(panel, output_name, order, xlabel):
    df = bar_data[bar_data['panel'] == panel].copy()
    fig, ax = plt.subplots(1, 1, figsize=(8, 7.5))
    sns.barplot(data=df, x='type', y='bias', order=order, capsize=0.2, palette=FOUR_COLORS, ax=ax)
    sns.stripplot(data=df, x='type', y='bias', order=order, hue='type', hue_order=order, palette=[sns.set_hls_values(color, l=0.35) for color in FOUR_COLORS], jitter=0.08, size=5, alpha=0.8, legend=False, edgecolor='black', linewidth=0.6, zorder=3, ax=ax)
    ax.set_ylim(1.0, 1.8)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6, 1.8, 2.0])
    ax.set_xlabel(xlabel)
    ax.set_ylabel('Level of Polarization')
    fig.savefig(OUTPUT_DIR / output_name, bbox_inches='tight', pad_inches=0.1, dpi=300)
    plt.show()

draw_polarization_bar('figure3e', 'figure3e_selective_exposure_bar.pdf', ['0%', '25%', '50%', '75%'], 'Ratio of Agents with Selective Exposure')
draw_polarization_bar('figure3f', 'figure3f_confirmation_bias_bar.pdf', ['0%', '25%', '50%', '75%'], 'Ratio of Agents with Confirmation Bias')
draw_polarization_bar('figure3g', 'figure3g_elite_agents_bar.pdf', ['Original', 'Neutral', 'Moderate', 'Radical'], 'Opinions of Elite Agents')
```
