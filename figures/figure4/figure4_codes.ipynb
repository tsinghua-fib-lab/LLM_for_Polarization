# Figure 4 codes

This notebook contains the plotting code for main-text Figure 4b, Figure 4c, Figure 4d, Figure 4e, and Figure 4f using the CSV files in `data/`.


```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

BASE_DIR = Path.cwd()
DATA_DIR = BASE_DIR / 'data'
OUTPUT_DIR = BASE_DIR / 'output'
OUTPUT_DIR.mkdir(exist_ok=True)

plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 25

PALETTE = ['#6384B3', '#F08743']
STRIP_PALETTE_BASE = '#355F8A'
STRIP_PALETTE_INTV = '#C45D1A'

PANELS = [
    ('figure4b', 'Random Interactions', 'RI'),
    ('figure4c', 'Moderate Opposing Interactions', 'MOI'),
    ('figure4d', 'No Selective Exposure', 'NSE'),
    ('figure4e', 'No Confirmation Bias', 'NCB'),
    ('figure4f', 'Neutral Elite Signaling', 'NES'),
]
```

```python
polarization = pd.read_csv(DATA_DIR / 'figure4_polarization_after_intervention.csv')
homophily = pd.read_csv(DATA_DIR / 'figure4_homophily_after_intervention.csv')
polarization.head(), homophily.head()
```

## Figure 4b-f


```python
def draw_horizontal_bar(df, panel, intervention_name, short_name, metric, xlabel, xlim, xticks, output_name):
    panel_df = df[df['panel'] == panel].copy()
    order = ['Original', intervention_name]
    strip_palette = {'Original': STRIP_PALETTE_BASE, intervention_name: STRIP_PALETTE_INTV}
    fig, ax = plt.subplots(1, 1, figsize=(10, 4))
    sns.barplot(data=panel_df, y='type', x=metric, order=order, capsize=0.2, palette=PALETTE, saturation=1, orient='h', ax=ax)
    sns.stripplot(data=panel_df, y='type', x=metric, order=order, orient='h', hue='type', palette=strip_palette, size=5, jitter=0.08, alpha=0.8, legend=False, edgecolor='black', linewidth=0.6, zorder=3, ax=ax)
    ax.set_yticks([0, 1])
    ax.set_yticklabels(['Orig', short_name])
    ax.set_xlim(*xlim)
    ax.set_xticks(xticks)
    ax.set_xlabel(xlabel)
    ax.set_ylabel('')
    fig.savefig(OUTPUT_DIR / output_name, bbox_inches='tight', pad_inches=0.1, dpi=300)
    plt.show()

for panel, intervention_name, short_name in PANELS:
    draw_horizontal_bar(
        polarization, panel, intervention_name, short_name,
        metric='polarization',
        xlabel='Polarization Level after Intervention',
        xlim=(1.0, 1.4),
        xticks=[1.0, 1.1, 1.2, 1.3, 1.4],
        output_name=f'{panel}_opinion_intervention_{short_name}.pdf',
    )
    draw_horizontal_bar(
        homophily, panel, intervention_name, short_name,
        metric='homophilic',
        xlabel='Homophilic Interactions after Intervention',
        xlim=(0, 0.6),
        xticks=[0, 0.2, 0.4, 0.6],
        output_name=f'{panel}_network_intervention_{short_name}.pdf',
    )
```
